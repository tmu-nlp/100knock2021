```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression

def yosoku(Logi):
    X_train=pd.read_table("X_train.txt")
    probs=(Logi.predict_proba(X_train))
    Cate=Logi.predict(X_train)
    max_pro=[]
    for num in probs:
        max_pro.append(max(num))
    for num2 in range(len(max_pro)):
        print(Cate[num2],"{:.4g}".format(max_pro[num2]))

LR=LogisticRegression(random_state=100,max_iter=50000)
df_train=pd.read_table("train.txt",header=None,names=["CATEGORY","TITLE"])
X_train=pd.read_table("X_train.txt")
LR.fit(X_train, df_train["CATEGORY"])

#53
yosoku(LR)
```
