```python
from collections import defaultdict
name_count=defaultdict(lambda:0)

with open("popular-names.txt","r") as name_file:
    for line in name_file:
        line=line.strip()
        words=line.split("\t")
        name=words[0]
        name_count[name]+=1
    sort_num=sorted(name_count.items(),reverse=True,key=lambda x:x[1])
    #print(sort_num)
    for line in sort_num:
        print(line[0],line[1])
    
#cut -f 1  popular-names.txt | sort | uniq -c | sort -r
```
