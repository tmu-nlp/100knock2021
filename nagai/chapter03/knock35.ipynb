```python
from  collections import Counter
with open("neko2.txt.mecab","r",encoding="utf-8") as me_file:
    keitaiso=[]
    kekka=[]
    sent=[]
    for line in me_file:
        line=line.strip("\n")
        line=line.replace(",","\t")
        items=line.split("\t")
        
        if items[0]=="EOS":
            break;
            
        keys={
            "surface":str(items[0]),
            "base":str(items[len(items)-3]),
            "pos":str(items[1]),
            "pos1":str(items[2]),
        }
        sent.append(keys)
        kekka.append(items[len(items)-3])
        if items[0]=="。":
            keitaiso.append(sent)
            sent=[]    
       
    freq=Counter(kekka)
    hindo=sorted(freq.items(),key=lambda x:x[1],reverse=True)
    for i in hindo:
        print(i)
```
