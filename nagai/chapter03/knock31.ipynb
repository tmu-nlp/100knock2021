```python
import MeCab
import pprint
with open("neko.txt","r",encoding="utf-8") as mecab_file:
    a=mecab_file.read()
    mecab=MeCab.Tagger()
with open("neko2.txt.mecab","w",encoding="utf-8") as mecab_file2:
    mecab_file2.write(mecab.parse(a))

with open("neko2.txt.mecab","r",encoding="utf-8") as me_file:
    keitaiso=[]
    sent=[]
    surf=[]
    for line in me_file:
        line=line.strip("\n")
        #line=line.replace(",*","")
        line=line.replace(",","\t")
        items=line.split("\t")
        
        if items[0]=="EOS":
            break;
            
        keys={
            "surface":str(items[0]),
            "base":str(items[len(items)-3]),
            "pos":str(items[1]),
            "pos1":str(items[2]),
        }
        sent.append(keys)
        if items[0]=="。":
            keitaiso.append(sent)
            sent=[]

    for line in keitaiso:
        for i in range(0,len(line)):
            if line[i]["pos"]=="動詞":
                surf.append(line[i]["surface"])
    print("\n".join(surf))
```
