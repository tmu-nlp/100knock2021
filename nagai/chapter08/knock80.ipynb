```python
import collections
def count(file):
    all_word=[]
    with open(file,"r",encoding="utf-8") as tf:
        for line in tf:
            line.strip();
            tag,sen=line.split("\t")
            words=sen.strip().split(" ")
            for word in words:
                all_word.append(word)
        freq=collections.Counter(all_word)
        freq_sort=sorted(freq.items(),key=lambda x:x[1],reverse=True)
        #print(freq_sort)
        word_number={w:i+1 for i,(w,f) in enumerate(freq_sort) if f>=2}
        return word_number

def id_v(file,word_id):
    with open(file,"r",encoding="utf-8") as tf:
        id=[]
        for line in tf:
            line.strip();
            tag,sen=line.split("\t")
            words=sen.strip().split(" ")
            sent_ids=[]
            for word in words:
                if word in word_id:
                    sent_ids.append(word_id[word])
                else:
                    sent_ids.append(0)
            id.append(sent_ids)
        return id
                       
word_id=count("train.feature.txt")
id_vector=id_v("train.feature.txt",word_id)
for sen in id_vector:
    print(sen)
```
