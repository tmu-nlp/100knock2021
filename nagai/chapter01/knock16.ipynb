```python
with open("kaisu.txt","r") as num:
    N=int(num.read())
    
with open("popular-names.txt","r") as name_file:
    lines=name_file.readlines()
    length=len(lines)
    line=int(length/N)
    for i in range(N):
        start=i*line
        print("".join(lines[start:start+line]))
        print("\n")
#split -n 10 -d popular-names.txt data        
```
