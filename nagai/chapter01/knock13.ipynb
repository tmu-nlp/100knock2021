```python
with open('col1.txt', 'r') as text1:
    with open('col2.txt', 'r') as text2:
        with open('merge.txt','w') as merge2:
            for word1,word2 in zip(text1,text2):
                w1=word1.rstrip('\n')
                merge="{0}\t{1}".format(w1,word2)
                print(merge)
                merge2.write(str(merge))
                
#paste col1.txt col2.txt > merge.txt
```
